```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line graph")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("pie chart")
plt.plot(x)
plt.show()
```

```python
x = [10,20,30,40]
plt.pie(x)
plt.show()
```

```python
x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
plt.pie(x, labels = y)
plt.show()
```

```python
x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
z = ["black" , "red" , "lightblue" , "b"]
plt.pie(x, labels = y, colors = z)
plt.show()
```

```python
x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
z = [0.2,0,0,0]
plt.pie(x, labels = y, explode = z)
plt.show()
```

```python
x = (["A" , "B" , "C" , "D"])
y = ([3,8,1,10])
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.scatter(x,y,color = "black")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
z = ["blue","red","black","grey","purple"]
plt.scatter(x,y,color = z)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
z = ["blue","red","black","grey","purple"]
plt.scatter(x,y,color = z,marker = "*")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
z = ["blue","red","black","grey","purple"]
plt.scatter(x,y,color = z,marker = "*",edgecolor = "green")
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
colors = ["blue","red","black","grey"]
sizes = [100,200,300,400]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker ="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=10)
plt.show()
```

```python
x = [1,2,3,4]
y = [4,5,6,7]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)

x = [10,20,30,40]
y = ["Apple" , "Banana" , "Mange" , "Orange"]
plt.pie(x, labels = y)
plt.subplot(2,2,3)

x = (["A" , "B" , "C" , "D"])
y = ([3,8,1,10])
plt.bar(x,y)
plt.subplot(2,2,4)

x = [1,2,3,4]
y = [1,2,3,4]
colors = ["blue","red","black","grey"]
sizes = [100,200,300,400]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker ="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=10)
```
